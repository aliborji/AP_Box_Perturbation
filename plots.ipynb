```python
from matplotlib import pyplot as plt
```

```python
# data_source = 'Ground Truth'
data_source = 'Model Prediction'

# perturbations = ['Random', 'Right', 'Left', 'Up', 'Down', 'Diag45', 'Diag-45']
# perturbations = ['scale']
# perturbations = ['shrink']
perturbations = ['Random', 'top-left', 'top-right', 'bottom-right', 'bottom-left', 'enlarge', 'shrink']

xx = 1.1

for pert in perturbations:
    
    if data_source.startswith('Model'):
        pert = 'm_' + pert
    
    # perturbation = f'{pert} direction'    
    perturbation = f'{pert}'        



    with open(f'{perturbation.split()[0].lower()}.txt', 'r') as f:
        data = f.readlines()
    

    perfs = []
    start = 9
    for k in range(10):
        d = data[start + 21*k]
        # print(d)
        perfs.append(float(d.strip()[-5:]))

    perfs_50 = []
    start = 10
    for k in range(10):
        d = data[start + 21*k]
        # print(d)
        perfs_50.append(float(d.strip()[-5:]))    

    perfs_75 = []
    start = 11
    for k in range(10):
        d = data[start + 21*k]
        # print(d)
        perfs_75.append(float(d.strip()[-5:]))        



    perfs_s = []
    start = 12
    for k in range(10):
        d = data[start + 21*k]
        # print(d)
        perfs_s.append(float(d.strip()[-5:]))

    perfs_m = []
    start = 13
    for k in range(10):
        d = data[start + 21*k]
        # print(d)
        perfs_m.append(float(d.strip()[-5:]))

    perfs_l = []
    start = 14
    for k in range(10):
        d = data[start + 21*k]
        # print(d)
        perfs_l.append(float(d.strip()[-5:]))   
        
        
        
    if data_source.startswith('Model'):
        perturbation = perturbation[2:]
        xx = .71
        

    plt.subplot(121)

    plt.plot(perfs, color = 'k', lw = 2, marker = 'o', markersize = 5)
    plt.plot(perfs_50, color = 'r', lw = 2, marker = 'o', markersize = 5)
    plt.plot(perfs_75, color = 'b', lw = 2, marker = 'o', markersize = 5)
    plt.legend(['mAP', 'AP50', 'AP75'])

    plt.xticks(range(10))
    plt.xlabel('pixel shift')
    plt.ylabel('AP')
    plt.yticks(np.arange(0,xx,.1))    
    plt.suptitle(f'          {data_source} Boxes  &  {perturbation.title()} Perturbation', fontweight ="bold")

    plt.grid(True)


    plt.subplot(122)
    plt.plot(perfs_s, color = 'k', lw = 2, linestyle='--', marker = 'o', markersize = 5)
    plt.plot(perfs_m, color = 'k', lw = 2, linestyle='-.', marker = 'o', markersize = 5)
    plt.plot(perfs_l, color = 'k', lw = 2, linestyle=':', marker = 'o', markersize = 5)

    plt.legend(['AP-S', 'AP-M', 'AP-L'])
    plt.xticks(range(10))
    plt.xlabel('pixel shift')
    # plt.ylabel('AP')
    plt.yticks(np.arange(0,xx,.1))
    # plt.set_yticklabels([])    

    # plt.title(f'{data_source} boxes  -  {perturbation.title()} pert.  (Obj Size)')
    plt.grid(True)


    plt.tight_layout()
    fName = ''.join(data_source.split()) + '-' + ''.join(perturbation.split())
    plt.savefig(f'{fName}.png', dpi=150)
    plt.show()

        
```
